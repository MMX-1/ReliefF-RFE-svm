```python
import pandas as pd
import numpy as np
tongren_data=pd.read_csv('tongren_600085.csv')
tongren_arr=np.array(tongren_data)
```

```python
#检查有无字符数
for i in range(0,51):
    print(i,tongren_arr[:,i])
```

```python
#处理字符串
a=tongren_arr[:,17]
num=len(a)
b=[]
flo=[]
#处理逗号
for i in range(num):
    
    b.append(a[i].replace(",",""))
    #print(str2float(i))
#str转换浮点数
for i in range(num):
    flo.append(float(b[i]))
#保存数据
for i in range(num):
    tongren_arr[i,17]=flo[i]
```

```python
#股票数量
stock_num=len(tongren_arr[:,1])
#目标变量
open_price_today=tongren_arr[:stock_num-1,1]#当日开盘价
open_price_yest=tongren_arr[1:stock_num,1]#昨日开盘价
list_=list(open_price_today-open_price_yest)
target=[]
#target
for i in list_:
    if i>=0:
        target.append(1)
    else:
        target.append(-1)
target=np.array(target)
```

```python
#标准化数据
fea=tongren_arr[1:,3:]
mean=np.mean(np.array(fea),axis=0,keepdims=True)
num_samp=fea.shape[0]
fea_num=fea.shape[1]
mean,fea,fea_num,num_samp
```

```python
#求方差
squa_s=np.sum((fea-mean)**2,axis=0)/(num_samp-1)#方差
std=[]
for i in range(fea_num):
    std.append(np.sqrt(squa_s[i]))
```

```python
#特征标准化
stan_s=np.array(std).reshape(1,fea_num)
norm_fea=(fea-mean)/stan_s
norm_fea
```

```python
X_y=np.c_[norm_fea,target]#将X和target合并
#导出数据
data_1=pd.DataFrame(X_y)
data_1.to_csv(r'norm_tongren.csv')
```
