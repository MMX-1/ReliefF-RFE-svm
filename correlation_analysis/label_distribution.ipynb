```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#% matplotlib inline


category_names = ['1', '-1']
train = {
    '603919': [jinhui_1_train,jinhui_minu1_train],
    '603369': [jinshi_1_train,jinshi_minu1_train],
    '600300': [weiwei_1_train,weiwei_minu1_train],
    "600559": [laobai_1_train,laobai_minu1_train],
    "600519": [guizhou_1_train,guizhou_minu1_train],
    "600809": [shanxi_1_train,shanxi_minu1_train],
    "600085": [tongren_1_train,tongren_minu1_train],
    "002304": [yanghe_1_train,yanghe_minu1_train],
    "600332": [baiyunshan_1_train,baiyunshan_minu1_train]
}

test = {
    '603919': [jinhui_1_test,jinhui_minu1_test],
    '603369': [jinshi_1_test,jinshi_minu1_test],
    '600300': [weiwei_1_test,weiwei_minu1_test],
    "600559": [laobai_1_test,laobai_minu1_test],
    "600519": [guizhou_1_test,guizhou_minu1_test],
    "600809": [shanxi_1_test,shanxi_minu1_test],
    "600085": [tongren_1_test,tongren_minu1_test],
    "002304": [yanghe_1_test,yanghe_minu1_test],
    "600332": [baiyunshan_1_test,baiyunshan_minu1_test]
}


def survey(train,test, category_names):
    """
    Parameters
    ----------
    results : dict
        A mapping from question labels to a list of answers per category.
        It is assumed all lists contain the same number of entries and that
        it matches the length of *category_names*.
    category_names : list of str
        The category labels.
    """
    labels_train = list(train.keys())
    train = np.array(list(train.values()))
    train_cum = train.cumsum(axis=1)
    labels_test = list(test.keys())
    test = np.array(list(test.values()))
    test_cum = test.cumsum(axis=1)
    category_colors = plt.get_cmap('RdYlGn')(
        np.linspace(0.15, 0.85, train.shape[1]))
    category_colors = plt.get_cmap('RdYlGn')(
        np.linspace(0.15, 0.85, test.shape[1]))
    fig, ((ax1,ax2)) = plt.subplots(figsize=(12,4),dpi=100,nrows=1,ncols=2)
    ax1.invert_yaxis()
    ax1.xaxis.set_visible(False)
    ax1.set_xlim(0, np.sum(train, axis=1).max())
    font1={'family': 'Times New Roman', 'weight': 'light', 'size': 10}#设置字体模板，
    
    for i, (colname, color) in enumerate(zip(category_names, category_colors)):
        widths = train[:, i]
        starts = train_cum[:, i] - widths
        ax1.barh(labels_train, widths, left=starts, height=0.5,label=colname,
                 color=color)
        xcenters = starts + widths / 2

        r, g, b, _ = color
        text_color = 'white' if r * g * b < 0.5 else 'darkgrey'
        for y, (x, c) in enumerate(zip(xcenters, widths)):
            ax1.text(x, y, str(int(c)), ha='center', va='center',
                    color=text_color)
    ax1.legend(ncol=len(category_names), bbox_to_anchor=(0, 1),
              loc='lower left', fontsize='xx-small')
    ax1.set_yticklabels(labels_train,fontdict=font1)
    ax1.set_title("Training Set")
    
    
    ax2.invert_yaxis()
    ax2.xaxis.set_visible(False)
    ax2.set_xlim(0, np.sum(test, axis=1).max())
    font1={'family': 'Times New Roman', 'weight': 'light', 'size': 10}#设置字体模板，
    
    for i, (colname, color) in enumerate(zip(category_names, category_colors)):
        widths = test[:, i]
        starts = test_cum[:, i] - widths
        ax2.barh(labels_test, widths, left=starts, height=0.5,label=colname,
                 color=color)
        xcenters = starts + widths / 2

        r, g, b, _ = color
        text_color = 'white' if r * g * b < 0.5 else 'darkgrey'
        for y, (x, c) in enumerate(zip(xcenters, widths)):
            ax2.text(x, y, str(int(c)), ha='center', va='center',
                    color=text_color)
    ax2.legend(ncol=len(category_names), bbox_to_anchor=(0, 1),
              loc="lower left", fontsize='xx-small')
    ax2.set_yticklabels(labels_train,fontdict=font1)
    ax2.set_title("Test Set")
    
    return fig, ax1,ax2


survey(train,test, category_names)
plt.savefig("label_distribution.eps",format="eps",dpi=1000, bbox_inches='tight')

```
